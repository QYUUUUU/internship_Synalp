```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch

quantization_config = BitsAndBytesConfig(
   load_in_4bit=True,
   bnb_4bit_compute_dtype=torch.bfloat16
)

model_name = "Qwen/Qwen2.5-Math-1.5B-Instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto",
    quantization_config=quantization_config,
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

print(model)
```

```python
tokenizer.pad_token_id = 151654
tokenizer.eos_token_id = 151643
```

```python
from datasets import load_dataset

dataset = load_dataset("openai/gsm8k","main", split='train')
dataset = dataset.train_test_split(test_size=0.003)
print(dataset)
```

```python
from trl import SFTConfig, SFTTrainer, DataCollatorForCompletionOnlyLM
import re

small_train = dataset["train"]
small_eval = dataset["test"]

def convert_to_boxed(answer):
    """
    Converts answer ending in '#### result' to '... \\boxed{result}' (with escaped backslash).
    """
    match = re.search(r'####\s*(\S+)', answer)
    if match:
        result = match.group(1)
        # Use double backslash to produce a literal \boxed{...} in the string
        answer = re.sub(r'####\s*\S+', f'\\\\boxed{{{result}}}', answer)
    return answer


def formatting_train_func(data):
    prompt = f'\nQuestion: {data["question"]} \nAnswer:'
    
    label = convert_to_boxed(data["answer"])
    
    messages = [
    {"role": "user", "content": prompt},
    {"role": "assistant", "content": label}
    ]


    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=False
    )

    return {"text": text}

small_train = small_train.map(formatting_train_func, remove_columns=["question", "answer"])
small_eval = small_eval.map(formatting_train_func, remove_columns=["question", "answer"])

print(small_train[0])
print(small_eval[0])
response_template = "<|im_start|>assistant\n"
collator = DataCollatorForCompletionOnlyLM(response_template, tokenizer=tokenizer)
```

```python
from peft import LoraConfig, TaskType, get_peft_model


# create LoRA configuration object
lora_config = LoraConfig(
    task_type=TaskType.SEQ_CLS, # type of task to train on
    inference_mode=False, # set to False for training
    r=8, # dimension of the smaller matrices
    lora_alpha=32, # scaling factor
    lora_dropout=0.01, # dropout of LoRA layers=
)

model.add_adapter(lora_config, adapter_name="lora_1")
```

```python
import numpy as np
from transformers import TrainingArguments, Trainer
import csv
import os
from trl import SFTConfig

class CustomTrainer(SFTTrainer):
    """
    A custom Trainer that logs the length of logits during loss computation.
    """
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        outputs = model(**inputs)
        
        labels = inputs.get("labels", None)
        attention_mask = inputs.get("attention_mask", None)
        logits = outputs.get('logits')
        log_entry = {"step": self.state.global_step}

        # Compute loss using parent class
        result = super().compute_loss(model, inputs, return_outputs=True)
        loss, output_values = result
        
        if labels is not None and attention_mask is not None:
            label_lengths = (attention_mask * (labels != -100)).sum(dim=1).tolist()
            log_entry["label_length"] = label_lengths[0]
        
        if logits is not None:
            predictions = torch.argmax(logits, dim=-1)
            pred_lengths = (predictions != tokenizer.pad_token_id).sum(dim=1).tolist()
            log_entry["predicted_answer_length"] = pred_lengths[0]

            # Convert predictions and labels to strings
            cleaned_predictions = predictions.clone()
            cleaned_predictions[cleaned_predictions == -100] = tokenizer.pad_token_id
            decoded_preds = tokenizer.batch_decode(cleaned_predictions, skip_special_tokens=True)

            
            cleaned_labels = labels.clone()
            cleaned_labels[cleaned_labels == -100] = tokenizer.pad_token_id
            decoded_labels = tokenizer.batch_decode(cleaned_labels, skip_special_tokens=True)
            log_entry["prediction"] = decoded_preds[0]
            log_entry["label"] = decoded_labels[0]

        if logits is not None and loss:
            log_entry["loss"] = loss.item()

        # Write to CSV
        with open("./logs/logging.csv", mode='a', newline='', encoding='utf-8') as f:
            writer = csv.DictWriter(f, fieldnames=["step", "loss", "label_length", "predicted_answer_length", "prediction", "label"])
            if f.tell() == 0:  # Write header only if file is new/empty
                writer.writeheader()
            writer.writerow(log_entry)

        return (loss, outputs) if return_outputs else loss


def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    sum_comparisons = 0
    answer_marker = tokenizer.encode("\\boxed{", add_special_tokens=False)
    end_marker = tokenizer.encode("<|im_end|>", add_special_tokens=False)[0]  # assume it's a single token
    length_eval_set = len(predictions)

    for i in range(length_eval_set):
        cutoff_index = np.argmax(labels[i] != -100)

        label_ids = labels[i][cutoff_index:]
        prediction_ids = predictions[i][cutoff_index:]

        # Find start of answer in label
        label_marker_pos = None
        for j in range(len(label_ids) - len(answer_marker) + 1):
            if list(label_ids[j:j + len(answer_marker)]) == answer_marker:
                label_marker_pos = j + len(answer_marker)
                break

        if label_marker_pos is not None:
            # Find end of label answer (where <|im_end|> appears)
            try:
                label_end_pos = np.where(label_ids == end_marker)[0][0]
            except IndexError:
                continue  # skip if <|im_end|> not found

            label_answer = label_ids[label_marker_pos:label_end_pos]
            
            # Find same start in prediction
            pred_marker_pos = None
            for j in range(len(prediction_ids) - len(answer_marker) + 1):
                if list(prediction_ids[j:j + len(answer_marker)]) == answer_marker:
                    pred_marker_pos = j + len(answer_marker)
                    break

            if pred_marker_pos is not None:
                pred_answer = prediction_ids[pred_marker_pos:pred_marker_pos + len(label_answer)]

                if np.array_equal(label_answer, pred_answer):
                    sum_comparisons += 1
    return {"exact_match_accuracy": (sum_comparisons / length_eval_set) * 100}


with open("./logs/logging.csv", mode='w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=["step", "loss", "label_length", "predicted_answer_length", "prediction", "label"])
    writer.writeheader()
    
sft_config = SFTConfig(
    output_dir="./data/working/maths_work/",
    num_train_epochs=1,
    logging_dir="./tensorboard/maths/logs",
    save_strategy="steps",
    eval_strategy="steps",
    push_to_hub=False,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    eval_accumulation_steps=1,
    gradient_accumulation_steps=1,
    gradient_checkpointing=True,
    save_total_limit=1,
    max_grad_norm=1.0,
    logging_steps=10,
    eval_steps=700,
    learning_rate=1.5e-5,
    report_to="tensorboard",
    load_best_model_at_end=False,
    max_seq_length= 1024,
)

trainer = CustomTrainer(
    model=model,
    train_dataset=small_train,
    eval_dataset=small_eval,
    args=sft_config,
    data_collator=collator,
    compute_metrics=compute_metrics,
)

trainer.train()

model.save_pretrained("./data/maths_ft/own_loop/model_weights.pth", from_pt=True)
```

```python
### from transformers import pipeline
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

dataset = load_dataset("nyu-mll/glue", "mnli", split='validation_matched[-100:]')
def evaluation_with_confusion_matrix(model):
    sum_scores = 0
    dict_results = {}

    y_true = []
    y_pred = []

    valid_labels = ["True", "False", "Neither"]

    for i in range(len(dataset)):
        prompt = f'{dataset[i]["premise"]} \nQuestion: {dataset[i]["hypothesis"]} \nTrue, False or Neither?\nAnswer:'

        messages = [
            {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ]

        chat_input = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True
        )

        inputs = tokenizer(chat_input, return_tensors="pt").to(model.device)

        with torch.no_grad():
            output = model.generate(
                **inputs,
                max_new_tokens=10,
                do_sample=False,
                #temperature=0.7,
                #pad_token_id=tokenizer.eos_token_id
            )
        decoded = tokenizer.decode(output[0], skip_special_tokens=False)        
        # Extract content after <|im_start|>assistant
        assistant_marker = "<|im_start|>assistant"
        if assistant_marker in decoded:
            assistant_part = decoded.split(assistant_marker)[-1].strip()
        else:
            assistant_part = decoded.strip()
        
        # Take only the first word and clean up
        first_word = assistant_part.split()[0] if assistant_part else ""
        first_word_cleaned = first_word.replace("<|im_end|>", "").strip()
        
        prediction = "Other"  # Default fallback
        for label in valid_labels:
            if first_word_cleaned.startswith(label):
                prediction = label
                break
        print("Prediction (cleaned):", prediction)
        
        observed_label = get_label(dataset[i]["label"]).strip()
        
        # Grade only if it's a valid response
        grade = 1 if prediction == observed_label else 0
        sum_scores += grade
        
        y_true.append(observed_label)
        y_pred.append(prediction)
        
        dict_results[i] = {
            "predicted": prediction,
            "observed": observed_label,
            "score": grade,
            "raw_output": assistant_part
        }


    # Accuracy
    accuracy = sum_scores / len(dataset)
    print(f"\n Accuracy: {accuracy:.3f}\n")

    # Confusion Matrix with "Other" class
    all_labels = valid_labels + ["Other"]
    cm = confusion_matrix(y_true, y_pred, labels=all_labels)

    plt.figure(figsize=(7,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=all_labels, yticklabels=all_labels)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix (with 'Other')")
    plt.show()

    # Classification Report
    print("\n Classification Report:\n")
    print(classification_report(y_true, y_pred, labels=all_labels, zero_division=0))

    return dict_results, sum_scores


results = evaluation_with_confusion_matrix(model)

quantization_config = BitsAndBytesConfig(
   load_in_4bit=True,
   bnb_4bit_compute_dtype=torch.bfloat16
)

model_name = "Qwen/Qwen2.5-0.5B-Instruct"


base_model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto",
)


results = evaluation_with_confusion_matrix(base_model)
```
